```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
math_df = pd.read_csv("dataset\student_math_clean.csv")
math_df.head()
```

```python
print([math_df['class_failures'].unique()])
```

```python
math_df.info()   #checking info
```

```python
math_df.describe()
```

```python
# missing values

math_df.isnull().sum()
```

Students whose final grade is 75% or better

```python
top_rank = math_df[math_df["final_grade"] >= 15]
top_rank.shape
```

```python
fig, axes = plt.subplots(6, 5, figsize=(25, 40))
axes = axes.flatten()

for i, col in enumerate(top_rank.columns[1:31]):
    sns.countplot(data=top_rank, x=col, ax=axes[i])
    axes[i].set_title(f'Countplot for {col}')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
low_rank = math_df[math_df['final_grade']<15]
low_rank.shape
```
